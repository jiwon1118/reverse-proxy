```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import pandas as pd
```

## 성능 그래프 그리기

```python
df_python_1200 = pd.read_csv("python_1200_output.csv")
df_numpy_1200 = pd.read_csv("numpy_1200_output.csv")
df_python_1300 = pd.read_csv("python_1300_output.csv")
df_numpy_1300 = pd.read_csv("numpy_1300_output.csv")
df_python_1200_40M = pd.read_csv("python_1200_40M_output.csv")
df_numpy_1200_40M = pd.read_csv("numpy_1200_40M_output.csv")
```

### python, numpy cache memory = 20M / vuser = 1200명 비교

http://172.17.0.1:8083/add-count-large-arrays-python

http://172.17.0.1:8083/add-count-large-arrays-numpy
- N = 10 ** 5 -> 10만개
- Cache (limit: cpus: "1.05" / memory: "20M")
- vuser = 1200
- Ramp-up 3000
- no limit

```python
# DateTime 열을 datetime 형식으로 변환
# x축: 시간 (DateTime)
x1 = pd.to_datetime(df_python_1200["DateTime"])
x2 = pd.to_datetime(df_numpy_1200["DateTime"])

# y축 데이터
y1_python = df_python_1200["Mean_Test_Time_(ms)"]
y2_python = df_python_1200["TPS"]

y1_numpy = df_numpy_1200["Mean_Test_Time_(ms)"]
y2_numpy = df_numpy_1200["TPS"]

# 전체 그래프 크기 설정
plt.figure(figsize=(14, 6))

# 첫 번째 그래프 (Python)
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째
plt.plot(x1, y1_python, color="blue", label="Mean_Test_Time(ms)")
plt.plot(x1, y2_python, color="red", label="TPS")
plt.xlabel("Time (DateTime)")
#plt.ylabel("")
plt.title("Python (1200vuser / memory=20M)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 두 번째 그래프 (NumPy)
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째
plt.plot(x2, y1_numpy, color="blue", label="Mean_Test_Time(ms)")
plt.plot(x2, y2_numpy, color="red", label="TPS")
plt.xlabel("Time (DateTime)")
#plt.ylabel("")
plt.title("Numpy (1200vuser / memory=20M)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 그래프 출력
plt.tight_layout()  # 그래프 간격 자동 조정
plt.show()
```
