## 성능 그래프 그리기

```python
df_python_1000 = pd.read_csv("python_1000_nocache_output.csv")
df_numpy_1000 = pd.read_csv("numpy_1000_nocache_output.csv")
df_python_1000_1mil = pd.read_csv("python_1000_1mil_output.csv")
df_numpy_1000_1mil = pd.read_csv("numpy_1000_1mil_output.csv")
df_python_700_1mil = pd.read_csv("python_700_1mil_output.csv")
df_numpy_700_1mil = pd.read_csv("numpy_700_1mil_output.csv")
```

- vuser = 1000

```python
# DateTime 열을 datetime 형식으로 변환
# x축: 시간 (DateTime)
x1 = pd.to_datetime(df_python_1000["DateTime"])
x2 = pd.to_datetime(df_numpy_1000["DateTime"])

# y축 데이터
y1_python = df_python_1000["Mean_Test_Time_(ms)"]
y2_python = df_python_1000["TPS"]

y1_numpy = df_numpy_1000["Mean_Test_Time_(ms)"]
y2_numpy = df_numpy_1000["TPS"]

# 전체 그래프 크기 설정
plt.figure(figsize=(14, 6))

# 첫 번째 그래프 (Python)
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째
plt.plot(x1, y1_python, color="blue", label="Mean_Test_Time(ms)")
plt.plot(x1, y2_python, color="red", label="TPS")
plt.xlabel("Time (DateTime)")
#plt.ylabel("")
plt.title("Python (1000vuser / no cache)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 두 번째 그래프 (NumPy)
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째
plt.plot(x2, y1_numpy, color="blue", label="Mean_Test_Time(ms)")
plt.plot(x2, y2_numpy, color="red", label="TPS")
plt.xlabel("Time (DateTime)")
#plt.ylabel("")
plt.title("Numpy (1000vuser / no cache)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 그래프 출력
plt.tight_layout()  # 그래프 간격 자동 조정
plt.show()
```

## output.csv와 stats.csv 파일 비교하기

- 10초를 기준으로 맞춘다
- 최대, 최소값을 이용해 시간을 맞춘다
- 차트를 그려본다

```python
df_python_stats = pd.read_csv("stats/python_700_1mil_docker_stats.csv")
df_numpy_stats = pd.read_csv("stats/numpy_700_1mil_docker_stats.csv")
```

```python
df_python_stats['Timestamp'] = pd.to_datetime(df_python_stats['Timestamp'], unit='s')
df_python_stats['Timestamp'] = df_python_stats['Timestamp'].dt.tz_localize('UTC').dt.tz_convert('Asia/Seoul').dt.tz_localize(None)

df_python_stats.head()
```

```python
df_numpy_stats['Timestamp'] = pd.to_datetime(df_numpy_stats['Timestamp'], unit='s')
df_numpy_stats['Timestamp'] = df_numpy_stats['Timestamp'].dt.tz_localize('UTC').dt.tz_convert('Asia/Seoul').dt.tz_localize(None)

df_numpy_stats.head()
```

```python
# df_python_700_1mil의 처음과 끝 Timestamp 가져오기
start_time = df_python_700_1mil["DateTime"].min()
end_time = df_python_700_1mil["DateTime"].max()

# 지정된 Timestamp 범위 내에서 'reverse-proxy-api-1' 필터링
df_filtered_python = df_python_stats[
    (df_python_stats["Timestamp"] >= start_time) &
    (df_python_stats["Timestamp"] <= end_time) &
    (df_python_stats["Name"] == "reverse-proxy-api-1")
]

x1 = pd.to_datetime(df_filtered_python["Timestamp"])

# y값을 숫자로 변환
df_filtered_python.loc[:, "CPUPerc"] = df_filtered_python["CPUPerc"].str.rstrip('%').astype(float)
df_filtered_python.loc[:, "MemPerc"] = df_filtered_python["MemPerc"].str.rstrip('%').astype(float)

y1_python = df_filtered_python["CPUPerc"]
y2_python = df_filtered_python["MemPerc"]


# df_numpy_700_1mil의 처음과 끝 Timestamp 가져오기
start_time_numpy = df_numpy_700_1mil["DateTime"].min()
end_time_numpy = df_numpy_700_1mil["DateTime"].max()

# 지정된 Timestamp 범위 내에서 'reverse-proxy-api-1' 필터링
df_filtered_numpy = df_numpy_stats[
    (df_numpy_stats["Timestamp"] >= start_time_numpy) &
    (df_numpy_stats["Timestamp"] <= end_time_numpy) &
    (df_numpy_stats["Name"] == "reverse-proxy-api-1")
]

x2 = pd.to_datetime(df_filtered_numpy["Timestamp"])

df_filtered_numpy.loc[:, "CPUPerc"] = df_filtered_numpy["CPUPerc"].str.rstrip('%').astype(float)
df_filtered_numpy.loc[:, "MemPerc"] = df_filtered_numpy["MemPerc"].str.rstrip('%').astype(float)


y1_numpy = df_filtered_numpy["CPUPerc"]
y2_numpy = df_filtered_numpy["MemPerc"]

# 전체 그래프 크기 설정
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# 첫 번째 그래프 (Python)
axs[0, 0].plot(x1, y1_python, color="blue", label="Python CPUPerc")
axs[0, 0].set_xlabel("Time (DateTime)")
axs[0, 0].set_ylabel("CPUPerc")
axs[0, 0].set_title("Python CPUPerc")
axs[0, 0].legend()
axs[0, 0].grid(True)
axs[0, 0].tick_params(axis='x', rotation=45)

# 두 번째 그래프 (Python MemPerc)
axs[0, 1].plot(x1, y2_python, color="red", label="Python MemPerc")
axs[0, 1].set_xlabel("Time (DateTime)")
axs[0, 1].set_ylabel("MemPerc")
axs[0, 1].set_title("Python MemPerc")
axs[0, 1].legend()
axs[0, 1].grid(True)
axs[0, 1].tick_params(axis='x', rotation=45)

# 세 번째 그래프 (NumPy CPUPerc)
axs[1, 0].plot(x2, y1_numpy, color="blue", label="NumPy CPUPerc")
axs[1, 0].set_xlabel("Time (DateTime)")
axs[1, 0].set_ylabel("CPUPerc")
axs[1, 0].set_title("NumPy CPUPerc")
axs[1, 0].legend()
axs[1, 0].grid(True)
axs[1, 0].tick_params(axis='x', rotation=45)

# 네 번째 그래프 (NumPy MemPerc)
axs[1, 1].plot(x2, y2_numpy, color="red", label="NumPy MemPerc")
axs[1, 1].set_xlabel("Time (DateTime)")
axs[1, 1].set_ylabel("MemPerc")
axs[1, 1].set_title("NumPy MemPerc")
axs[1, 1].legend()
axs[1, 1].grid(True)
axs[1, 1].tick_params(axis='x', rotation=45)

# 그래프 출력
plt.tight_layout()  # 그래프 간격 자동 조정
plt.show()
```

```python
# df_numpy_700_1mil의 처음과 끝 Timestamp 가져오기
start_time_numpy = df_numpy_700_1mil["DateTime"].min()
end_time_numpy = df_numpy_700_1mil["DateTime"].max()

# 지정된 Timestamp 범위 내에서 'reverse-proxy-api-1' 필터링
df_filtered_numpy = df_numpy_stats[
    (df_numpy_stats["Timestamp"] >= start_time_numpy) &
    (df_numpy_stats["Timestamp"] <= end_time_numpy) &
    (df_numpy_stats["Name"] == "reverse-proxy-api-1")
]

x = pd.to_datetime(df_filtered_numpy["Timestamp"])

df_filtered_numpy.loc[:, "CPUPerc"] = df_filtered_numpy["CPUPerc"].str.rstrip('%').astype(float)
df_filtered_numpy.loc[:, "MemPerc"] = df_filtered_numpy["MemPerc"].str.rstrip('%').astype(float)


y1_numpy = df_filtered_numpy["CPUPerc"]
y2_numpy = df_filtered_numpy["MemPerc"]

# 전체 그래프 크기 설정
plt.figure(figsize=(14, 6))

# 첫 번째 그래프 (Python)
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째
plt.plot(x, y1_numpy, color="blue", label="numpy CPUPerc")
plt.xlabel("Time (DateTime)")
plt.ylabel("CPUPerc")
plt.title("numpy CPUPerc")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 두 번째 그래프 (NumPy)
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째
plt.plot(x, y2_numpy, color="red", label="numpy MemPerc")
plt.xlabel("Time (DateTime)")
plt.ylabel("MemPerc")
plt.title("numpy MemPerc")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# 그래프 출력
plt.tight_layout()  # 그래프 간격 자동 조정
plt.show()
```
